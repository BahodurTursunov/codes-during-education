```python
import requests as req
from bs4 import BeautifulSoup
from time import sleep
import random
from numpy import random
import pandas as pd
```

```python
from selenium.webdriver import Chrome
driver = Chrome(executable_path='chromedriver')
```

```python
data = []


for page in range(1, 11):
    sleep(random.uniform(3, 7))
    url = f"https://www.kinopoisk.ru/lists/top500/?page={page}&tab=all"
    driver.get(url)
    html = driver.page_source
    print(page)
    s = BeautifulSoup(html, "lxml")
    
    films = s.findAll('div', class_='desktop-rating-selection-film-item')
    
    for film in films:
        link = "https://www.kinopoisk.ru"+film.find('a', class_='selection-film-item-meta__link').get('href')
        rus_title = film.find('a', class_='selection-film-item-meta__link').find('p', class_='selection-film-item-meta__name').text
        orig_title = film.find('a', class_='selection-film-item-meta__link').find('p', class_='selection-film-item-meta__original-name').text
        country = film.find('a', class_='selection-film-item-meta__link').findAll('span', class_='selection-film-item-meta__meta-additional-item')[0].text
        genre = film.find('a', class_='selection-film-item-meta__link').findAll('span', class_='selection-film-item-meta__meta-additional-item')[1].text
        rate = film.find('span', 'rating__value rating__value_positive').text
    
        data.append([link, rus_title, orig_title, country, genre, rate])
```

```python
head = ['link', 'rus_title', 'orig_title', 'country', 'genre', 'rate']
dt = pd.DataFrame(data, columns=head)
dt.to_csv('kinopoisk_data.csv', sep=';', encoding='utf-16')
```
